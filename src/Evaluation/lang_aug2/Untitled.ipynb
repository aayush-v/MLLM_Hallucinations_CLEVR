```python
import numpy as np
import json
import os
```

```python
def calculate_precision_recall(tp, fp, tn, fn):
    precision = tp / (tp + fp) if (tp + fp) != 0 else 0
    recall = tp / (tp + fn) if (tp + fn) != 0 else 0
    f1 = 2 * precision * recall / (precision + recall)
    return precision, recall, f1
```

```python
def evaluate(path):
    files = os.listdir(path)
    for file in files:
        consistency_results_path = os.path.join(path, file)
        if not os.path.isfile(consistency_results_path):
            continue
        f = open(consistency_results_path,)
        objects = json.load(f)
    
    
        yes_tp = 0
        yes_fp = 0
        yes_fn = 0
        yes_tn = 0
        for object in objects:
            if object['ground_truth'] == 'yes' and object['consistent'] == 'yes':
                yes_tp += 1
            elif object['ground_truth'] == 'yes' and object['consistent'] == 'no':
                yes_fn += 1
            elif  object['ground_truth'] == 'no' and object['consistent'] == 'yes':
                yes_tn += 1
            elif object['ground_truth'] == 'no' and object['consistent'] == 'no':
                yes_fp += 1
    
        no_tp = 0
        no_fp = 0
        no_fn = 0
        no_tn = 0
        for object in objects:
            if object['ground_truth'] == 'no' and object['consistent'] == 'yes':
                no_tp += 1
            elif object['ground_truth'] == 'no' and object['consistent'] == 'no':
                no_fn += 1
            elif  object['ground_truth'] == 'yes' and object['consistent'] == 'yes':
                no_tn += 1
            elif object['ground_truth'] == 'yes' and object['consistent'] == 'no':
                no_fp += 1
    
        print(file, len(objects))
        print(yes_tp, yes_fp, yes_tn, yes_fn)
        print((yes_tp+yes_fn)/(yes_tn+yes_fp))
        print(calculate_precision_recall(yes_tp, yes_fp, yes_tn, yes_fn))
        # print()
        print(no_tp, no_fp, no_tn, no_fn)
        print(calculate_precision_recall(no_tp, no_fp, no_tn, no_fn))
        print((no_tp+no_fn)/ (no_tn+no_fp))
        print()
        print()
```
